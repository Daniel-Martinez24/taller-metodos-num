```
# import
import numpy as np
import matplotlib.pyplot as plt

n=100
i0= 1
beta = .003
def i(i0, beta, t): return n*i0/(i0-(i0-n)*np.exp(-n*beta*t))
def s(*args): return n - i(*args)
```

```
t_array = np.arange(0,50,1)
i_array = i(i0, beta, t_array) # formula para infercados
s_array = s(i0, beta, t_array) # para susptibles
print(t_array)
print(s_array)
```

```
plt.plot(t_array, i_array, label = 'infectados')
plt.plot(t_array, s_array, label = 'susceptibles')
plt.grid(True)
plt.legend()
plt.show()
```

```
fig = plt.figure(figsize=(16,8))
for color, beta in zip(['r', 'b', 'k', 'g'],[.01, .005, .004, .002]):
    t_array = np.arange(0, 50, 1)
    i_array = i(i0, beta, t_array)
    s_array = s(i0, beta, t_array)
    plt.plot(t_array, i_array, c = color, label = 'infectados beta = {} '.format(beta))
plt.grid(True)
plt.legend(fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.show()
```
